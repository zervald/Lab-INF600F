## Exercice 1
Essayer de détecter des coins/caractéristiques intéressant avec la méthode Harris. Testez avec différent seuil.

```python
img = data.brick()
print(img.shape, img.min(), img.max())

plt.figure(figsize=(20, 5))

plt.subplot(141); plt.imshow(feature.corner_harris(img), cmap="gray"); plt.title('img')

fig = feature.corner_harris(img )
s1 = fig > 0.0
plt.subplot(142)
plt.imshow(s1)
plt.title("0.0")

s2 = fig > 0.01
plt.subplot(143)
plt.imshow(s2)
plt.title("0.01")

s3 = fig > 0.03
plt.subplot(144)
plt.imshow(s3, cmap="gray")
plt.title("0.03")

plt.show()
```

## Exercice 2
Appliquer un filtre détecteur de forme pour faire ressortir les contours continus (lignes/tubes) dans l'image.
