Complétez la fonction `compare` pour comparer 2 nombres.

```python
def compare(nb1, nb2):
    """
    Si nb1 < nb2, retourne -1
    Si nb1 > nb2, retourne 1
    Si nb1 == nb2, retourne 0
    """
    comparaison = None
    
    # Si nb1 < nb2, retourne (-1, 'inferieur')
    if nb1 < nb2:
        return (-1, 'inferieur')
    
    # Si nb1 > nb2, retourne (1, 'superieur')
    if nb1 > nb2:
        return 1
    
    # Si nb1 == nb2, retourne (0, 'egal')
    if nb1 == nb2:
        return 0
    
    return comparaison

a = 10; b = 25
print('{:<4}'.format(str(a)), '{:<4} :'.format(str(b)), compare(a, b))
a = 111; b = 92
print('{:<4}'.format(str(a)), '{:<4} :'.format(str(b)), compare(a, b))
a = 1000; b = 1000
print('{:<4}'.format(str(a)), '{:<4} :'.format(str(b)), compare(a, b))
```

## Observations

```python
# Python fait usage de `deep` et `shallow copy` dépendemment du type de la variable passé.

print('Passé en paramètre un integer:')
a = 5

def add(nb, add):
    nb += add
    return nb

print(a)
print(add(a, 5))
print(a)

print('---')

print('Passé en paramètre une liste (objet mutable):')
list_a = [5, 10]

def list_add(list_nb, add):
    for i in range(0, len(list_nb)):
        list_nb[i] += add
    return list_nb

print(list_a)
print(list_add(list_a, 5))
print(list_a)

print('---')
print('Erreur typique que je fais souvent:')
test = [[0, 0, 0]] * 3
print(test)
test[0][0] = 1
print(test)

print('---')
print('Solution pour cette erreur:')
test = []

for i in range(0,3):
    test.append([0, 0, 0])

print(test)
test[0][0] = 1
print(test)
```

## Image to gray
ex image

```python
from skimage import io
from skimage import color
import numpy as np
import matplotlib.pyplot as plt

image = io.imread("image.jpg")
imageAsArray = np.array(image)
print(imageAsArray.shape)

def image2gray(image):
    if image.shape[2] == 4:
        imageRGB = color.rgba2rgb(image)
        return color.rgb2gray(imageRGB)
    else:
        return color.rgb2gray(image)

plt.subplot(1,2,1)
plt.imshow(imageAsArray)

plt.subplot(1,2,2)
plt.imshow(image2gray(imageAsArray))
plt.show
```
